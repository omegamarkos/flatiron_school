```python
import pandas as pd
import numpy as np

from sklearn.datasets import make_circles,  make_blobs, make_moons
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from mlxtend.plotting import plot_decision_regions
```

```python
# make data 
x, y = make_blobs(n_samples=500, n_features=2, centers=2, cluster_std=3.5)
```

```python
# plot data 
color_dict = {0: 'black', 1: 'blue'}
for data, label in zip(x, y):
    plt.scatter(data[0], data[1], c = color_dict[label])
```

```python
# euclidean_distance 
ecu = KNeighborsClassifier(p = 2)
ecu.fit(x, y)
print(ecu.score(x, y))
plot_decision_regions(x, y, ecu, colors = 'black,blue')
```

```python
# manhattan distance
manh = KNeighborsClassifier(p = 1)
manh.fit(x, y)
print(manh.score(x, y))
plot_decision_regions(x, y, manh, colors = 'black,blue')
```
