#### Edges - represent the relationships between these entities

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
import networkx as nx
```

```python
df_graph = nx.from_pandas_edgelist(df, source ='Origin', target = 'Dest', edge_attr = True)
```

```python
df_graph.nodes()
```

```python
df_graph.edges()
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.xkcd()
```

```python
plt.figure(figsize = (12, 8))
nx.draw_networkx(df_graph)
```

```python
x = dict(nx.degree(df_graph))
```

```python
{k: v for k, v in sorted(x.items(), key=lambda item: item[1], reverse = True)}
```

```python
df.head()
```
