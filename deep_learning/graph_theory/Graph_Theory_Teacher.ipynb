```python
df = pd.read_csv('data.csv')
df.head()
```

```python
import networkx as nx
df_graph = nx.from_pandas_edgelist(df, source='Origin', target='Dest', edge_attr=True)
```

```python
df_graph.nodes()
```

```python
df_graph.edges()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
nx.draw_networkx(df_graph, with_labels=True)
```

```python
df_graph.edges.values
```

```python
dict(nx.degree(df_graph))
```

```python
x = dict(nx.degree(df_graph))
```

```python
{k: v for k, v in sorted(x.items(), key=lambda item: item[1], reverse = True)}
```

```python
df.head()
```

```python
shortest_path_distance = nx.dijkstra_path(df_graph, source='MCI', target='LAS', weight='Distance')
shortest_path_distance
```

```python
shortest_path_distance = nx.dijkstra_path(df_graph, source='MCI', target='TUL', weight='Distance')
shortest_path_distance
```
